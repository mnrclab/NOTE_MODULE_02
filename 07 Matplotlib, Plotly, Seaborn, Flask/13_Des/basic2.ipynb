## **Heatmap**

```python
df = pd.read_csv('dataSales.csv')
df = df.pivot(index='nama', columns='mobil')
df
```

```python
fig, p = plt.subplots()
plt.imshow(df, cmap='hot')
plt.colorbar()

col = list(map(lambda x: x[1], df.columns.tolist()))
i = list(map(lambda x: x, df.index.tolist()))

p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)

p.set_yticks(np.arange(len(i)))
p.set_yticklabels(i)

plt.show()
```

```python
fig, p = plt.subplots()
plt.imshow(df, cmap='BuPu_r') #MEMBALIK DATANYA
plt.colorbar()

col = list(map(lambda x: x[1], df.columns.tolist()))
i = list(map(lambda x: x, df.index.tolist()))

for x in range(len(i)):
    for y in range(len(col)):
        p.text(y, x, df.iloc[x, y],
                       ha="center", va="center", color="k")

p.set_xticks(np.arange(len(col)))
p.set_xticklabels(col)

p.set_yticks(np.arange(len(i)))
p.set_yticklabels(i)

p.set_ylim(2.5,-0.5)
plt.show()
```

## **Correlation Matrix**

```python
data = pd.read_csv('atlit.csv')
dtCor = data.corr()
```

```python
plt.imshow(dtCor, cmap = 'BuPu_r')
plt.colorbar()

col = list(map(lambda x: x, dtCor.columns.tolist()))

plt.xticks(
    np.arange(len(dtCor.index)),
    dtCor.index.tolist(),
    rotation=45
)

plt.yticks(np.arange(len(dtCor.index)),
          dtCor.index.tolist())

for x in range(len(col)):
    for y in range(len(col)):
        plt.text(y, x, round(dtCor.iloc[x, y], 2),
                       ha="center", va="center", color="k")
plt.ylim(-.5, 4.5)
plt.show()
```

## **3D Plotting**

```python
fig = plt.figure()
p = plt.subplot(111, projection='3d')
data = range(11)
x = np.array(data)

p.plot_wireframe(
    x, x, x.reshape(1,-1), color='r', linestyle=':'
    )

p.scatter(x,x,x, marker='o', color='y', s=150)
p.set_xlabel('Sumbu X')
p.set_ylabel('Sumbu Y')
p.set_zlabel('Sumbu Z')

plt.show()
```
