## **Horizontal Bar**

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.barh(df['x'], df['y'], 
        color=['r','g','b','y','k'], #memberi warna yang berbeda di tiap kolom
        edgecolor='black', linewidth=2.5)
plt.title('Tes Bar Chart', fontdict={
    'family':'Impact', 'size': 15, 'color':'lightblue'
})
plt.xlabel('Sumbu X', size=10)
plt.ylabel('Sumbu Y', size=10)
plt.show()
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.bar(df['x'], df['y'], 
        color=['r','g','b','y','k'], #memberi warna yang berbeda di tiap kolom
        edgecolor='black', linewidth=2.5)
plt.yticks(np.arange(0,12,2))
plt.title('Tes Bar Chart', fontdict={
    'family':'Impact', 'size': 15, 'color':'lightblue'
})

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.1, i+.2, i,
            fontdict={'family': 'Calibri', 'size': 12}
            )

plt.xlabel('Sumbu X', size=10)
plt.ylabel('Sumbu Y', size=10)
plt.show()
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.bar(df['x'], df['y'], 
        color=['r'],
        edgecolor='black', linewidth=2.5)

plt.bar(df['x'], df['z'], 
        color=['b'],
        edgecolor='black', linewidth=2.5)


plt.yticks(np.arange(0,12,2))
plt.title('Tes Bar Chart', fontdict={
    'family':'Impact', 'size': 15, 'color':'lightblue'
})

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.1, i+.2, i,
            fontdict={'family': 'Calibri', 'size': 12}
            )

plt.xlabel('Sumbu X', size=10)
plt.ylabel('Sumbu Y', size=10)
plt.show()
```

```python
df = pd.DataFrame({
    'x': [1,2,3,4,5],
    'y': [9,5,7,3,6],
    'z': [2,3,4,5,3]
})

plt.bar(df['x'], df['y'], 
        color=['r'],
        edgecolor='black', linewidth=2.5)

plt.bar(df['x'], df['z'], 
        color=['b'], bottom=df['y'], yerr=.5, #memberi posisi di bawahnya nilainya df['y']
        edgecolor='black', linewidth=2.5)


plt.yticks(np.arange(0,12,2))
plt.title('Tes Bar Chart', fontdict={
    'family':'Impact', 'size': 15, 'color':'lightblue'
})

for i in df['y']:
    plt.text(df[df['y'] == i]['x']-.1, i+.2, i,
            fontdict={'family': 'Calibri', 'size': 12}
            )

plt.xlabel('Sumbu X', size=10)
plt.ylabel('Sumbu Y', size=10)
plt.show()
```

## **Pie Chart**

```python
persentase = [25,60,15]
keterangan = ['Pak A', 'Pak B','Pak C']
plt.pie(
    persentase, labels=keterangan,
    colors=['b','r', 'g'],
    startangle = 0, #untuk memulai chart, arahnya berlawanan arah jarum jam, 0,45,90,120
    shadow = True, #untuk menambahkan background
#     counterclock = False, #untuk membuat searah jarum jam
    textprops = {'size':15, 'color':'k'},
#     autopct = '%i%%' #kalau pengen tampil integer
    autopct = '%1.1f%%' #kalau pengen tampil float
)
plt.show()
```

```python
persentase = [25,60, 15]
keterangan = ['Pak A', 'Pak B','Pak C']

x, y , z = plt.pie(
    persentase, labels=keterangan,
    colors=['b','r', 'g'],
    textprops = {'size':15, 'color':'k'},
    autopct = '%1.1f%%',
    radius = .8,
    explode = (.1,0,.3) #supaya salah satu potongan keluar
)

for i in z:
    i.set_color('w')

plt.show()
```

## **Histogram**

```python
usia = [22,41,23,44,65,26,57,28,32,33,70,25,56,69]
usia2 = [32,31,33,34,55,36,47,38,22,23,64,27,58,63]
par = np.arange(15,70,5)


plt.hist(
    [usia, usia2], par, histtype='barstacked',
    rwidth=10
        )

plt.xticks(np.arange(15,70,5))
plt.show()
```
