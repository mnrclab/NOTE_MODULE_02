### **Groupby in Pandas**

```python
import pandas as pd
import numpy as np
```

```python
# Membuat dataframe
data = {'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB','FB'],
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]
       }
df = pd.DataFrame(data)
df
```

```python
df.groupby('Company')
```

```python
by_comp = df.groupby('Company')
by_comp.mean()
```

```python
df.groupby('Company').mean()
```

```python
by_comp.std()
```

```python
by_comp.count()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_person = df.groupby('Person')
by_person.mean()
```

```python
by_person.max()
```

```python
by_person.min()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_person.describe().transpose()['Amy']
```

```python
by_comp.describe().transpose()['FB']
```

```python
by_comp.describe().transpose()['GOOG']['Sales']['25%']
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'd':['D0','D1','D2','D3']},
                   index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'd':['D0','D1','D2','D3']},
                   index=[4, 5, 6, 7])
df3 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'd':['D0','D1','D2','D3']},
                   index=[8, 9, 10, 11])
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis=1)
```

```python
pd.concat([df1,df2,df3]).transpose()
```
