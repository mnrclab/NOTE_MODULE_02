```python
import numpy as np
import pandas as pd
```

## A. Mengakses Data Excel

**1) Membaca File Excel**

```python
ps = pd.read_excel('playstore.xlsx')
ps.head(2)
# ps.dtypes
```

**2) Menjadikan Last Updated sebagai Index**

```python
ps = ps.set_index('Last_Updated')
```

## B. Analisis Data

**1) Menampilkan Rata-rata Rating App dari tahun 2011-2018**

```python
ps['2011':'2018']['Rating'].resample('Y').mean()
```

**2) Menampilkan Jumlah Review Tiap Minggu di Bulan Desember tahun 2017**

```python
ps['2017-12']['Reviews'].resample('W').sum()
```

**2) Menampilkan APP (+category dan rating) yang memiliki rating tertinggi**

```python
ps[['App','Category','Rating']][
    ps['Rating'] == ps['Rating'].max()
]
```

```python
olah = pd.crosstab(ps.Category,[ps.Content_Rating, ps.Genres], margins=True, margins_name='Total')
olah
```

```python
ps[['Content_Rating','Category','Rating']].melt(id_vars=['Content_Rating','Category'], var_name='parameter', value_name='nilai')
```

```python
ps1 = ps[['Content_Rating','Genres','Rating']].melt(id_vars=['Content_Rating','Genres'], var_name='parameter', value_name='Rating').groupby(['Content_Rating','Genres'])
ps1.mean().head(20)
```
